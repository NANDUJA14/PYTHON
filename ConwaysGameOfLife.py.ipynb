```python
import numpy as np
import matplotlib.pyplot as plt
import time

def update(grid):
    new_grid = grid.copy()
    rows, cols = grid.shape

    for i in range(rows):
        for j in range(cols):
            # Count live neighbors
            neighbors = np.sum(grid[max(0, i-1):min(rows, i+2), max(0, j-1):min(cols, j+2)]) - grid[i, j]

            # Apply rules
            if grid[i, j] == 1 and (neighbors < 2 or neighbors > 3):
                new_grid[i, j] = 0  # Underpopulation or overpopulation
            elif grid[i, j] == 0 and neighbors == 3:
                new_grid[i, j] = 1  # Reproduction

    return new_grid

def display(grid):
    plt.imshow(grid, cmap="gray")
    plt.axis("off")
    plt.show(block=False)
    plt.pause(0.1)
    plt.clf()

def game_of_life(rows=20, cols=20, iterations=50):
    grid = np.random.choice([0, 1], size=(rows, cols))  # Random initial state

    plt.figure(figsize=(6, 6))

    for _ in range(iterations):
        display(grid)
        grid = update(grid)
        time.sleep(0.1)

    plt.show()

if __name__ == "__main__":
    game_of_life()
```
